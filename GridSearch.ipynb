## ***GridSearch***

```python
import pandas as pd
import numpy as np
np.random.seed(2)
random_state=2
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
digits = load_digits()
X_digits = digits.data
y_digits = digits.target

X_train_digits, X_test_digits, y_train_digits, y_test_digits = train_test_split(X_digits, y_digits, random_state=random_state)

param_grid_digits = [
    {'kernel': ['linear'], 'C': [0.1, 1, 10, 100]},
    {'kernel': ['rbf'], 'C': [0.1, 1, 10, 100], 'gamma': [0.001, 0.01, 0.1, 1]}
]

svm_clf_digits = SVC()

svm_grid_search_digits = GridSearchCV(svm_clf_digits, param_grid_digits, cv=5, scoring='accuracy', verbose=2)
svm_grid_search_digits.fit(X_train_digits, y_train_digits)
```

```python
svm_grid_search_digits.best_score_
```

```python
svm_grid_search_digits.best_params_
```

```python
best_svm_estimator = svm_grid_search_digits.best_estimator_
```

```python
y_pred = best_svm_estimator.predict(X_test_digits)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test_digits, y_pred)
print(f"Accuracy on test set: {accuracy:.4f}")
```

## ***RandomizedSearch***

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, uniform, randint

param_dist = {
    'C': uniform(0.1, 10),
    'gamma': uniform(0.001, 1),
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree': randint(2, 5)
}

random_search = RandomizedSearchCV(svm_clf_digits, param_distributions=param_dist, n_iter=20, cv=5, random_state=42, verbose=2)
random_search.fit(X_train_digits, y_train_digits)
```

```python
print("Best Parameters:", random_search.best_params_)
print("Best Score:", random_search.best_score_)

best_model = random_search.best_estimator_
test_score = best_model.score(X_test_digits, y_test_digits)
print("Test Set Score:", test_score)
```
